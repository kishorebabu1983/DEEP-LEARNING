```python
img_gray=cv2.imread("images/test_images/flowers/f1.jpg",0)  # 0 is for gray scale, 1 for BGR
img_color=cv2.imread("images/test_images/flowers/f1.jpg",1)
```

```python
cv2.imshow("color image",img_color)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#cv2 represents images as BGR instead of RGB
```

```python
plt.imshow(img_color)
```

```python
plt.imshow(cv2.cvtColor(img_color,cv2.COLOR_BGR2RGB))
```

```python
#READ LIST OF ALL IMAGES
import cv2
import glob
```

```python
path="images/test_images/flowers/*.*"
for file in glob.glob(path):
#     print(file)
    img=cv2.imread(file,1)
#     print(img)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    cv2.imshow("img",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
